```python
import pandas as pd
from transformers import (
    AutoModelForSeq2SeqLM,
    AutoModelForTableQuestionAnswering,
    AutoTokenizer,
    pipeline,
)
```

```python
model_tapas = "google/tapas-base-finetuned-sqa"
tokenizer_tapas = AutoTokenizer.from_pretrained(model_tapas)
model_tapas= AutoModelForTableQuestionAnswering.from_pretrained(model_tapas)
pipe_tapas = pipeline(
    "table-question-answering", model=model_tapas, tokenizer=tokenizer_tapas
)
```

```python
import pandas as pd

df = pd.read_csv("/content/sample_doc.csv", encoding="latin-1")
df = df.astype(str).head(12)
print(df)
```

```python
df.dropna(inplace=True)
```

```python
query = "what is the customer no?"
result_tapas = pipe_tapas(table=df, query=query)
result_tapas
```
